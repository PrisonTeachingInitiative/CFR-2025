#### Example 4: Histogram

Histograms are useful for displaying the distribution of a dataset.


```python
# Generate our data
#***********************
# Step 1: Generate 100 random numbers
data = np.random.randn(1000)
data[:11]
# Step 2: Create Dataframe with the data
df = pd.DataFrame(data, columns=["Values"])
df.head(10)
#Step 3: save dataframe to  csv file (Can be used in excel)

name_csv_file = "data_session10_example4.csv"
df.to_csv(name_csv_file, index=False)

# Step4: Bring in the data from CSV.
df_from_csv = pd.read_csv(name_csv_file)
df_from_csv.head(10)

# Create plot
#*************************************
plt.hist(df_from_csv["Values"], bins=30, edgecolor='purple')
plt.title("Histogram of Values")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid()
```
