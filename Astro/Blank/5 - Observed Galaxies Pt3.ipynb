Let's calculate the percent of each type of galaxy in the two data groups. Remember, we can calculate the percent of something by doing: $\rm \frac{number\ of\ specific}{total} \times 100$

Calcualte and compare the 4 percents below.

```python
# first we need to bring in the data you created in the last notebook. copy and paste the lines of code to read in that data

import pandas as pd
data_class = pd.read_csv('../data/galazyzoo_experiment.csv')

data_class
```

```python
# copy and paste ALL lines of code after you printed the data table. (You can ignore the print statements at the end.)

# For my data (data1)
elliptical = data1["P_ELLIPTICAL"]
spiral = data1["P_SPIRAL"]

elliptical_galaxy = elliptical >= 0.8
spiral_galaxy = spiral >= 0.8

total_classified = (elliptical_galaxy | spiral_galaxy).sum()
print('The total number of galaxies  classified is', total_classified)

# My data percentages
num_my_elliptical = elliptical_galaxy.sum()
num_my_spiral = spiral_galaxy.sum()
total_my = len(data1)

percent_my_elliptical = (num_my_elliptical / total_my) * 100
percent_my_spiral = (num_my_spiral / total_my) * 100

# For Galaxy Zoo (data_class)
tot_num_classifications = data_class["Total # of classifications"]
elliptical_all = data_class["Elliptical"]
spiral_all = data_class["Spiral"]

prob_elliptical = elliptical_all / tot_num_classifications
prob_spiral = spiral_all / tot_num_classifications

is_elliptical = prob_elliptical >= 0.8
is_spiral = prob_spiral >= 0.8

num_all_elliptical = is_elliptical.sum()
num_all_spiral = is_spiral.sum()
total_all = len(data_class)

percent_all_elliptical = (num_all_elliptical / total_all) * 100
percent_all_spiral = (num_all_spiral / total_all) * 100

# Final output
print('For data1:')
print('The percentage of elliptical galaxies is', percent_my_elliptical, '%')
print('The percentage of spiral galaxies is', percent_my_spiral, '%')
print('  ')
print('For all of the Galaxy Zoo data:')
print('The percentage of elliptical galaxies is', percent_all_elliptical, '%')
print('The percentage of spiral galaxies is', percent_all_spiral, '%')
```

Let's think about the ages of spiral galaxies and elliptical galaxies. It is hard to know the age of a galaxy, so we can focus on general trends.

Below add your code to test your hypothesis. Have a clearly labeled plot that you are ready to present to the group. Add statistics to your findings by exploring the mean and standard deviation of your results.

```python
# add your code below to test your hypothesis

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import matplotlib
matplotlib.rc('font', **{'family':'sans-serif', 'size':16}) 

# Load the data
data1 = pd.read_csv('../Data/galaxyzoo_research.csv')

# Filter by galaxy type probability
ellipticals = data1[data1["P_ELLIPTICAL"] > 0.8]
spirals = data1[data1["P_SPIRAL"] > 0.8]

# Compute color indices
elliptical_galaxy_color = ellipticals["GMAG"] - ellipticals["RMAG"]
spiral_galaxy_color = spirals["GMAG"] - spirals["RMAG"]

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(spiral_galaxy_color, bins=25, alpha=0.6, label='Spirals', color='skyblue', edgecolor='black')
plt.hist(elliptical_galaxy_color, bins=30, alpha=0.6, label='Ellipticals', color='red', edgecolor='black')


plt.xlabel('G - R (Color Index)', fontsize=14)
plt.ylabel('Number of Galaxies', fontsize=14)
plt.title('Color Distribution of Spiral and Elliptical Galaxies', fontsize=16)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Statistics
print('The mean color of spiral galaxies is', round(np.mean(spiral_galaxy_color), 2))
print('The mean color of elliptical galaxies is', round(np.mean(elliptical_galaxy_color), 2))
print()
print('The standard deviation of the color of spiral galaxies is', round(np.std(spiral_galaxy_color), 2))
print('The standard deviation of the color of elliptical galaxies is', round(np.std(elliptical_galaxy_color), 2))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import matplotlib
matplotlib.rc('font', **{'family':'sans-serif', 'size':16}) 

# Load the data
data1 = pd.read_csv('../Data/galaxyzoo_research.csv')

# Filter by galaxy type probability
ellipticals = data1[data1["P_ELLIPTICAL"] > 0.8]
spirals = data1[data1["P_SPIRAL"] > 0.8]

# Compute age proxies: U - I
elliptical_galaxy_age = ellipticals["UMAG"] - ellipticals["IMAG"]
spiral_galaxy_age = spirals["UMAG"] - spirals["IMAG"]

# Plot histograms for age
plt.figure(figsize=(10, 6))
plt.hist(spiral_galaxy_age, bins=25, alpha=0.6, label='Spirals', color='green', edgecolor='black')
plt.hist(elliptical_galaxy_age, bins=30, alpha=0.6, label='Ellipticals', color='orange', edgecolor='black')

plt.xlabel('U - I (Age Indicator)', fontsize=14)
plt.ylabel('Number of Galaxies', fontsize=14)
plt.title('Age Distribution of Spiral and Elliptical Galaxies (U - I)', fontsize=16)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Statistics
print('The mean U - I of spiral galaxies is', round(np.mean(spiral_galaxy_age), 2))
print('The mean U - I of elliptical galaxies is', round(np.mean(elliptical_galaxy_age), 2))
print()
print('The standard deviation of U - I for spiral galaxies is', round(np.std(spiral_galaxy_age), 2))
print('The standard deviation of U - I for elliptical galaxies is', round(np.std(elliptical_galaxy_age), 2))
```
