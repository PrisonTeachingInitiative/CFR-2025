## Magnitudes

We use "magnitudes" to describe how bright a star is. Remember, we mentioned that stars give off a lot of light. That light makes it to Earth, and we can measure how bright the star is from that light. The magnitude system was developed a long time ago, and unforunately, it is a weird scale. The higher a star's magnitude, the DIMMER it is. Bright stars have a low magnitude. (This is backwards from how I would personally do it! I try to remember this scale by saying the magnitude is the number of telescopes you'd need to see the star. If you need a high number, it must be dim.)

There's another thing that is special about star light. It is a lot of different kinds of light! Humans can see the "visible light," and we feel what's called "infrared light" has heat. While our eyes can only process visible light, we can use machines to "see" the other kind of light coming from a star. We use "filters" to limit the light to a certain kind. This is similar to how wearing sunglasses block out light!

## Color

Color is another thing astronomers define a little weirdly. We are used to seeing a color that our eyes can pick up. That color corresponds to a certain kind of light.

Astronomers define color by subtracting the magnitudes in two different filters. For example, we could do g - r. That will give us a number, not what most people would call a color. We can say colors that are bigger than 0.6 are redder. And those that are smaller than 0.6 are bluer.

```python
# calculate color and save the values to a variable
color = gband - rband
  
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

# variable data
color = gband - rband
temp = data["Temperature [K]"]

#def
def inversely_proportional(x, A):
    return A / x

mask = color > 0
x = color[mask]  
y = temp[mask]    

# fit
popt, pcov = curve_fit(inversely_proportional, x, y)

# prediction
x_model = np.linspace(x.min(), x.max(), 1000)
y_model = inversely_proportional(x_model, *popt)

from sklearn.metrics import mean_squared_error

# Predict values using your fitted model
y_pred = inversely_proportional(x, *popt)

# Compute RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"RMSE of inverse model: {rmse:.2f}")
plt.figtext(1.02, 0.85, f"Fit Model:\nT = A / (g - r)\nA = {popt[0]:.2f}\nRMSE = {rmse:.2f}",
            fontsize=12, ha='left', va='top', bbox=dict(facecolor='white', edgecolor='gray'))

# data model
plt.scatter(color, temp, c=temp, cmap='plasma', alpha=0.5, label='Data')
plt.plot(x_model, y_model, color='black', linestyle='--', linewidth=2, label='Fit: T = A / (g - r)')

plt.title("Color vs Temperature (with Fit)", fontsize=18)
plt.xlabel("g - r Color", fontsize=14)
plt.ylabel("Temperature [K]", fontsize=14)

plt.xlim(color.min() - 0.1, color.max() + 0.1)
plt.ylim(temp.min() - 200, temp.max() + 200)

plt.axvline(0.6, color='black', linestyle='--')
plt.text(0.3, temp.max() * 0.95, 'Bluer', color='blue', fontsize=12)
plt.text(0.7, temp.max() * 0.95, 'Redder', color='red', fontsize=12)

plt.legend()
plt.colorbar(label='Temperature [K]')
plt.rc('font', **{'family': 'sans-serif', 'size': 18})
plt.grid(True)
plt.show()
```
